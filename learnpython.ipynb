```python
all_airports = []

with open('airports.csv', 'r') as airport_file:
    airport_file.readline() # Skip header line
    for airport in airport_file:
        all_airports.append(airport)

all_airports = all_airports[:len(all_airports) - 1]
```

```python
for airport in all_airports:
    print(airport, end='')
```

```python
for airport in all_airports:
    items = airport.split('","')
    airport_code = items[0][1:]
    print(airport_code)
```

```python
for airport in all_airports:
    # Split the airport code from the airport location and name and
    # remove the quotation mark from the beginning of the airport code
    items = airport.split('","')
    airport_code = items[0][1:]

    # Split the airport location and airport name, and remove the quotation
    # mark and newline character from the end of the airport name
    subitems = items[1].split(': ')
    airport_location = subitems[0]
    airport_name = subitems[1][:-2]

    # Print the resulting strings
    print(airport_code, airport_location, airport_name)
    #print('{0:8}{1:32}{2:1}'.format(airport_code, airport_location, airport_name))
```

```python
airports = []

for airport in all_airports:
    items = airport.split('","')
    airport_code = items[0][1:]
    subitems = items[1].split(': ')
    airport_location = subitems[0]
    airport_name = subitems[1][:-2]
    airports.append([airport_code, airport_location, airport_name])
    
print(len(airports))

for airport in airports:
    print(airport)
    #print(airport[0], airport[1], airport[2])
    #print('{0:8}{1:32}{2:1}'.format(airport[0], airport[1], airport[2]))
```

```python
us_airports = []
not_us_airports = []

us_states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA",
             "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD",
             "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ",
             "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC",
             "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]

for airport in airports:
    if airport[1].endswith(tuple(us_states)):
        us_airports.append(airport)
    else:
        not_us_airports.append(airport)
        
print(len(airports))
print(len(us_airports))
```

```python
for airport in not_us_airports:
    print(airport)
```

```python
def get_airports_by_state(airports, state):
    result = []
    for airport in airports:
        if airport[1].endswith(state):
            result.append(airport)
    return result
```

```python
state = input('Enter a state abbreviation: ')
state_airports = get_airports_by_state(us_airports, state)

print('Number of airports: ' + str(len(state_airports)))

for airport in state_airports:
    print(airport)
```
